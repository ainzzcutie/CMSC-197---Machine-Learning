```python
# import standard libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the data
df = pd.read_csv("Advertising.csv", index_col = 0)
print("Size of the data:", df.shape)
df.head()
```

```python
# standardize each column using sklearn's scale function
scaled_data = df.copy()
scaled_data[['TV', 'Radio', 'Newspaper', 'Sales']] = scale(df[['TV', 'Radio', 'Newspaper', 'Sales']])

print(scaled_data.head())
```

```python
scaled_data_with_bias = scaled_data.copy()
# insert the bias column with values 1
scaled_data_with_bias.insert(0, 'bias', 1)
print(scaled_data_with_bias.head())
```

```python
# old means and standard deviations to destandardize later on
TV_mean = df.TV.mean()
TV_std =  df.TV.std()
radio_mean =  df.Radio.mean()
radio_std =  df.Radio.std()
newspaper_mean =  df.Newspaper.mean()
newspaper_std =  df.Newspaper.std()
sales_mean =  df.Sales.mean()
sales_std =  df.Sales.std()
```

```python
# get independent variable
x = scaled_data_with_bias.iloc[:, :-1]
print(x.head())
```

```python
# get dependent variable
y = scaled_data_with_bias.iloc[:, -1]
print(y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)
```

```python
# functions
def de_standardize(scaled_values, mean, std_dev):
    de_standardized_values = scaled_values * std_dev + mean
    return de_standardized_values
    
def initialize_weights():
    return np.array([0.0, 0.0, 0.0, 0.0])

def predict(x, w):
    return np.dot(x, w)

def compute_cost(x, y, w):
    m = len(y)
    y_hat = predict(x, w)
    cost = (1/(2*m)) * np.sum((y_hat - y) ** 2)
    return cost

def compute_gradient(x, y, w):
    m = len(y)
    y_hat = predict(x, w)
    gradient = (1/m) * np.dot(x.T, (y_hat - y))
    return gradient

def update_weights(w, gradient, learning_rate):
    return w - gradient * learning_rate

def grad_descent(x, y, learning_rate, num_iterations):
    weights = initialize_weights()
    cost_history = []
    
    for i in range(num_iterations):
        cost = compute_cost(x, y, weights)
        cost_history.append(cost)
        
        gradient = compute_gradient(x, y, weights)
        weights = update_weights(weights, gradient, learning_rate)
    return weights, cost_history

def plot_costs(cost_history, learning_rate):
    plt.figure(figsize=(18, 8)) 
    plt.scatter(
        x = range(len(cost_history)),
        y = cost_history
    )
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Cost vs Iterations, Learning Rate={}'.format(learning_rate))
    plt.tight_layout()
    plt.show()
```

```python
# parameters
learning_rate = 0.01
num_iterations = 1000

# run gradient descent
weights, cost_history = grad_descent(x_train, y_train, learning_rate, num_iterations)
print("weights: ", weights)
print("Last 10 costs: ", cost_history[-10:])
```

```python
plot_costs(cost_history, learning_rate)
```

```python
# Predict y for training set and calculate cost
y_train_pred = predict(x_train, weights)
train_cost = compute_cost(x_train, y_train, weights)

# Predict y for test set and calculate cost
y_test_pred = predict(x_test, weights)
test_cost = compute_cost(x_test, y_test, weights)

# Compute R2 score
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
```

```python
# Revert the values of y_train_pred and y_train to non-standardized values
y_train_pred_reverted = de_standardize(y_train_pred, sales_mean, sales_std)
y_train_reverted = de_standardize(y_train, sales_mean, sales_std)


# Revert the values of y_test_pred and y_test to non-standardized values
y_test_pred_reverted = de_standardize(y_test_pred, sales_mean, sales_std)
y_test_reverted = de_standardize(y_test, sales_mean, sales_std)
```

```python
# display train comparison
train_comparison_df = pd.DataFrame({'Actual': y_train_reverted, 'Predicted': y_train_pred_reverted})

print("\nTrain Set Predictions:")
train_comparison_df.head()
```

# 3
#### What happens to the error, r2, and cost as the number of iterations increase?

```python
iteration_list = [100, 500, 1000, 2000, 5000, 10000, 25000, 50000]

for n_iter in iteration_list:
    weights, cost_history = grad_descent(x_train, y_train, learning_rate, n_iter)
    y_train_pred = predict(x_train, weights)
    y_test_pred = predict(x_test, weights)

    # Get the final cost (last value in cost_history)
    final_cost = cost_history[-1]
    
    # Print the cost and R² scores for the current iteration count
    print(f"Iterations: {n_iter}")
    print(f"Final Cost: {final_cost}")
    print(f"R^2 Score for {n_iter} iterations (Train): {r2_score(y_train, y_train_pred)}")
    print(f"R^2 Score for {n_iter} iterations (Test): {r2_score(y_test, y_test_pred)}\n")
```

#### Cost
The final cost decreases significantly in the early iterations (from 0.1057 at 100 iterations to 0.0513 at 500 iterations). However, after around 500 iterations, the cost stabilizes, with no significant reduction beyond that point. After 500 iterations, the cost value remains virtually constant at **0.051336**.

#### R2 (Train)
The R² score on the training set increases rapidly up to 500 iterations (from 0.783 to 0.893), indicating that the model is improving its fit to the training data. Beyond 500 iterations, there is no noticeable improvement in the R² score for the training set, which stabilizes at around **0.8937**.

#### R2 (Test)
The R² score on the test set follows a similar pattern: it improves significantly up to 500 iterations (from 0.765 to 0.909). After 500 iterations, the R² score on the test set continues to improve slightly, but by 1000 iterations and beyond, it reaches a plateau at approximately **0.9110**.

#### The optimal number of iterations can be determined by finding the point where the model converges, and further iterations offer diminishing returns. Based on the results:

After 1000 iterations, both the cost and R² scores stabilize, with very minimal improvement beyond that point.

At 1000 iterations, the training R² is **0.8937** and the test R² is **0.9110**, which is a strong fit for both the training and test data.

Further increasing the iterations up to 50,000 does not improve the cost or the R² scores, meaning that additional iterations are computationally wasteful without any significant benefit.

The optimal number of iterations is around **1000**. Beyond this point, the model converges, and additional iterations do not improve the performance significantly.

# 4
#### the effect on the cost and error as you change the learning rate..

```python
learning_rates = [0.1, 0.01, 0.001, 0.0001, 0.2]
iterations = 1000

plt.figure(figsize=(18, 8))
for alpha in learning_rates:
    weights, cost_history = grad_descent(x_train, y_train, alpha, iterations)
    plt.plot(cost_history, label=f"alpha = {alpha}")

plt.title("Cost over Iterations for Different Learning Rates (Scaled Dataset)")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()
```

Higher learning rates, such as 0.1, 0.2, or even 0.01, allow the model to converge more rapidly but can cause oscillations or even divergence if set too high. In contrast, lower learning rates, such as 0.0001, result in slower but more gradual convergence. As demonstrated in the plot, models with lower learning rates may not reach a stable value even by the 2000th iteration, while those with higher learning rates have already achieved convergence before this point. This highlights the trade-off between speed and stability in gradient descent optimization, where higher learning rates facilitate faster convergence but carry the risk of instability, whereas lower learning rates promote steady convergence at the cost of slower progress.


# 5
#### Is there a relationship on the learning rate and the number of iterations? 

There exists an inverse relationship between the learning rate and the number of iterations required for convergence. Higher learning rates reduce the number of iterations needed to reach convergence, but they come with the risk of overshooting the optimal solution or causing instability. On the other hand, lower learning rates require a greater number of iterations to converge, but they tend to offer more stable and reliable convergence. This trade-off highlights the importance of selecting an appropriate learning rate to balance between convergence speed and model stability.

Gradient Descent also performs better on the test set, with a significantly lower MSE compared to OLS.
