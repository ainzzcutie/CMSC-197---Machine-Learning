```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt

# Step 1: Load the Advertising.csv dataset using pandas
df = pd.read_csv('Advertising.csv')

# Extract predictors (TV, Radio, Newspaper) and response (Sales)
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

# Step 2: Standardize the columns of the dataset
X_scaled = scale(X)

# Add a column of ones for the bias term
X_scaled = np.c_[np.ones(X_scaled.shape[0]), X_scaled]  # Shape (n_samples, 4)

# Step 3: Split the dataset into training and testing sets
def split_data(X, y, train_ratio=0.85):
    n_train = int(train_ratio * len(X))
    X_train, X_test = X[:n_train], X[n_train:]
    y_train, y_test = y[:n_train], y[n_train:]
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = split_data(X_scaled, y)

# Step 4: Initialize weights (for 4 parameters: 1 bias, 3 predictors)
def initialize_weights():
    init_w = np.array([0.0, 0.0, 0.0, 0.0])
    return init_w

# Step 5: Predict function
def predict(X, w):
    return np.dot(X, w)

# Step 6: Compute cost (MSE function)
def compute_cost(X, y, w):
    m = len(y)
    predictions = predict(X, w)
    cost = (1/(2*m)) * np.sum((predictions - y) ** 2)
    return cost

# Step 7: Compute gradients
def compute_gradient(X, y, w):
    m = len(y)
    predictions = predict(X, w)
    gradient = (1/m) * np.dot(X.T, (predictions - y))
    return gradient

# Step 8: Update weights using gradient descent
def update_weights(w, grad, alpha):
    return w - alpha * grad

# Step 9: Gradient descent algorithm
def grad_descent(X, y, alpha, n_iterations):
    w = initialize_weights()  # Set seed for reproducibility
    cost_history = []

    for i in range(n_iterations):
        grad = compute_gradient(X, y, w)
        w = update_weights(w, grad, alpha)
        cost = compute_cost(X, y, w)
        cost_history.append(cost)
        
        # Optionally print progress
        # if i % 100 == 0:
        print(f"Iteration {i}: Cost {cost}")

    return w, cost_history

# Step 10: Plot costs
def plot_costs(cost_history):
    plt.plot(range(len(cost_history)), cost_history, 'b')
    plt.title("Cost Function over Iterations")
    plt.xlabel("Iteration")
    plt.ylabel("Cost")
    plt.show()

# Step 11: Run gradient descent
alpha = 0.05  # Learning rate
n_iterations = 1000

# Run gradient descent on training data
w_final, cost_history = grad_descent(X_train, y_train, alpha, n_iterations)

# Plot the cost history
plot_costs(cost_history)

# Step 12: Predict and calculate cost for train set
train_predictions = predict(X_train, w_final)
train_cost = compute_cost(X_train, y_train, w_final)
print(f"Train cost: {train_cost}")

# Step 13: Predict and calculate cost for test set
test_predictions = predict(X_test, w_final)
test_cost = compute_cost(X_test, y_test, w_final)
print(f"Test cost: {test_cost}")

# Step 14: Reverting the standardized predictions to original form
def revert_standardized_data(scaled_values, original_data):
    std_dev = np.std(original_data)
    mean_val = np.mean(original_data)
    return scaled_values * std_dev + mean_val

# Example: Reverting standardized train predictions
train_predictions_original = revert_standardized_data(train_predictions, df['Sales'])
print(train_predictions_original)
```
